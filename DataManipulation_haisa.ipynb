```python
### Importing Continent data ###
continent = pd.read_excel('continent.xlsx')
continent.set_index('Country',inplace=True)
continent.head()
```

```python
### Joining dataframes and Dealing with missing Continent Values! ###
df = df.join(continent, on='Country', how='left', lsuffix='Country', rsuffix='Country')
### Finding The Countrys without Continent data ### 
empty_cont = df[df['Continent'].isnull()]
empty_cont.index
```

```python
### Filling empty Continents using a dict ###
cont_dict = {'Hong Kong, China (SAR)': 'Asia','United Kingdom' : 'Europe','United States':'North America',
             'Korea (Republic of)':'Asia','Moldova (Republic of)':'Europe','Palestine, State of':'Asia',
             'Eswatini (Kingdom of)':'Africa', "CÃ´te d'Ivoire":'Africa',
             'Tanzania (United Republic of)':'Africa','Congo (Democratic Republic of the)':'Africa'}

cont_dict = pd.DataFrame(list(cont_dict.items()),columns = ['Country','Continent'])
cont_dict.set_index('Country',inplace=True)

df[df['Continent'].isnull()] = df[df['Continent'].isnull()].fillna(cont_dict)
```
