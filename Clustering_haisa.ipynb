```python
import pandas as pd
```

```python
df = pd.read_excel('dados_hdi.xlsx')
#df.head()
```

```python
df.set_index('Country',inplace=True)
#df.head()
#df.drop('Index', axis=1)
#df.describe()
#df.info()
```

```python
cols = df.columns.drop(['Index'])
cols
df[cols] = df[cols].apply(pd.to_numeric, errors = 'coerce')
#df.info()
```

```python
continent = pd.read_excel('continent.xlsx')
#continent.head()
continent.set_index('Country',inplace=True)
continent.head()
```

```python
######Dealing with missing Continent Values!
empty = df.columns[df.isna().any()].tolist()
df_cols = df[empty]
df_cols = df_cols.join(continent, on='Country', how='left', lsuffix='Country', rsuffix='Country')

cont = df_cols[['Continent']] 
empty_cont = cont[cont['Continent'].isnull()]
empty_cont
```

```python
###Filling empty Continents!!!
cont_dict = {'Hong Kong, China (SAR)': 'Asia','United Kingdom' : 'Europe','United States':'North America',
             'Korea (Republic of)':'Asia','Moldova (Republic of)':'Europe','Palestine, State of':'Asia',
             'Eswatini (Kingdom of)':'Africa', "Côte d'Ivoire":'Africa',
             'Tanzania (United Republic of)':'Africa','Congo (Democratic Republic of the)':'Africa'}

cont_dict = pd.DataFrame(list(cont_dict.items()),columns = ['Country','Continent'])
cont_dict.set_index('Country',inplace=True)
cont_dict
```

```python
cont = cont.fillna(cont_dict)
#cont.head()
```

```python
df = df.join(cont, on='Country', how='left', lsuffix='Country', rsuffix='Country')
df.head()
```

```python
#Verifing the columns with empty values!!!
empty = df.columns[df.isna().any()].tolist()
df_cols = df[empty]
df_cols
```

```python
##Calculating the means for each Continent!
df_cols_mean = df.groupby(['Continent']).mean()
df_cols_mean
```

```python
############ FILLING THE MISSING VALUES ######
###Finding them!!!
df_full = df

#### Dropping 2 collumns, Index and other that was duplicate!!!####
df_full = df_full.drop(['Gender Inequality Index Value (2019).1', 'Index'], axis=1)

#df.head()
df_full.isnull().sum()
```
